```python
#Importamos las dependencias
import pandas as pd
```

```python
#Este código utiliza pandas para leer un archivo CSV desde una URL en Google Sheets y lo carga en un DataFrame de pandas.
#Se establece la columna "Fecha" como el índice del DataFrame.
datos = pd.read_csv("https://docs.google.com/spreadsheets/d/e/2PACX-1vSJ6cHSZdTtEqhHCx7RpRXaMjxGU1swsfTxTKBkbdRT4Pg8KMXc5NNeYpTWIqpxjn-sJ05jYtImi1cU/pub?output=csv",index_col="Fecha")
```

```python
#Imprimos los datos que tenemos en el csv
datos
```

```python
#convierte todas las columnas de un DataFrame llamado datos a tipo numérico.
datos = datos.apply(pd.to_numeric, errors='coerce')
```
